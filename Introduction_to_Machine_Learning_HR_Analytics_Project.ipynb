## 2. Data Cleaning & Analysis

```python
# Loading the required libraries



import pandas as pd

import numpy as np


import warnings

warnings.filterwarnings('ignore')

```

```python
#read dataset
hr_df=pd.read_csv("https://bit.ly/2ODZvLCHRDataset")

hr_df.head()
```

```python
#check null values
hr_df.isnull().sum() 
```

```python

hr_df.drop(['employee_id'], axis=1, inplace=True)

#hr_df['education'] = hr_df['education'].fillna(hr_df['education'].mode(), inplace=True)
hr_df = hr_df[hr_df['education'].notna()]

hr_df['previous_year_rating'].fillna(hr_df['previous_year_rating'].median(), inplace=True)
```

```python
#one hot encoding

hr_df= pd.get_dummies(hr_df)
hr_df.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

target = hr_df['is_promoted']
features = hr_df.drop(['is_promoted'], axis=1)
features_train, features_valid, target_train, target_valid = train_test_split(
    features, target, test_size=0.25, random_state=12345
)
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(random_state=12345, n_estimators=3)
model.fit(features_train, target_train)
predicted_valid = model.predict(features_valid)


```

```python
new_hr_df=pd.read_csv("/content/test_2umaH9m.csv")
new_hr_df.drop(['employee_id'], axis=1, inplace=True)
new_hr_df['previous_year_rating'].fillna(new_hr_df['previous_year_rating'].median(), inplace=True)
new_hr_df= pd.get_dummies(new_hr_df)
new_hr_df.isnull().sum()

new_hr_df.head()

```
